```
import pandas as pd
import statsmodels.formula.api as sms
import matplotlib.pyplot as plt
import numpy as np

boston = pd.DataFrame.from_csv("https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv")

credit = pd.read_csv("http://www-bcf.usc.edu/~gareth/ISL/Credit.csv")
credit = credit.drop(columns=['Unnamed: 0'])
```

<h1>Exercício 1</h1>

```
model_fit1 = sms.ols(formula="medv~lstat", data=boston).fit()
model_fit2 = sms.ols(formula="medv~age", data=boston).fit()
model_fit3 = sms.ols(formula="medv~age+lstat", data=boston).fit()
print(model_fit1.summary(), model_fit2.summary(), model_fit3.summary())
```

modelo com lstat explica mais da feature medv (r2), embora ambos sejam altamente relacionados a medv (p-value) 
ambas features tem uma correlação negativa em relação a medv, ou seja quanto menor age e lstat, maior medv

```
model_fit4 = sms.ols(formula="medv~lstat+age+zn+indus+chas+nox+rm+dis+rad+tax+ptratio+b", data=boston).fit()
print(model_fit4.summary())
```

criando um modelo com todas as variáveis é possível notar que age, diferentemente do que visto em um modelo simples, não influencia medv (p-value > 0.05), bem como indus. ambas podem ser removidas do modelo sem perda de acurácia

```
model_fit5 = sms.ols(formula="medv~lstat+zn+chas+nox+rm+dis+rad+tax+ptratio+b", data=boston).fit()
print(model_fit5.summary())
```

sem age e indus r2 ajustado melhorou levemente 0.01%

<h1>Exercício 2</h1>

```
model_fit = sms.ols(formula="Balance~Gender", data=credit).fit()
print(model_fit.summary())
```

p-value maior que 0.05 <br>
não é possível rejeitar a hipótese nula, logo não há associação entre Balance e Gender<br>
demais coeficientes não foram sequer analisados

```
model_fit = sms.ols(formula="Balance~Ethnicity", data=credit).fit()
print(model_fit.summary())
```

p-value maior que 0.05 <br>
não é possível rejeitar a hipótese nula, logo não há associação entre Balance e Gender<br>
demais coeficientes não foram sequer analisados

```
model_fit = sms.ols(formula="Balance~Income+Limit+Rating+Cards+Age+Education+Gender+Student+Married+Ethnicity", data=credit).fit()
print(model_fit.summary())
```

p-value: maior que 0.05, para diversas features, existem relações onde podemos descartar hipótese nula<br>
As features Student, Income, Limit e Cards são as mais correlatas e que mais influenciam no modelo.<br>
R-squared ajustado: valor alto, significa que boa parte dos dadosé explicado pelo modelo. 95%<br>
Slope positivo para a maior parte das features consideradas relevantess(p-value = 0)<br>
significa que na medida que feature x cresce, Balance também cresce. Exemplo: em média, para cada USD 100 a mais de Limit, Balance cresce USD 19. Exceção: Income. para cada USD 1 a mais de Income, Balance diminui USD 8 <br>
<br>
Tiraria as features Education, Gender, Age, Married e Ethnicity, que não explicam praticamente nada dos dados

```
model_fit = sms.ols(formula="Balance~Income+Limit+Rating+Cards+Age+Student", data=credit).fit()
print(model_fit.summary())
```

Mesmo tirando todos as features cujo p-value é maior que 0.05, R-squared ajustado permanece 0.954, o mesmo com todas as variaveis. Tentativas de retirar Rating e Age (features cujo p-value sáo maiores que 0.000 pioraram o modelo) <br>
hipótese: Existe uma variabilidade nos dados que não pode ser explicada ou não faz parte das features disponibilizadas
